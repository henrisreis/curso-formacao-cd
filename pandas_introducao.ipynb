```python
import pandas as pd
```

```python
# Carrega arquivo para dataframe Pandas

dados = pd.read_csv("Credit.csv")

# Formato

dados.shape # Método que mostra a estrutura dos dados. 1000 linhas e 21 colunas.
```

```python
# Resumo estatístico das colunas numéricas

dados.describe()
```

```python
# Para visualizar os primeiros registros. Ideal para observar a estrutura dos dados.

dados.head()
```

```python
# Para visualizar os 2 últimos registros

dados.tail(2)
```

```python
# Para visualizar os dados de uma coluna específica

dados[["duration"]]
```

```python
# Para exibir as linhas de índices 1 a 3

dados.loc[1:3]
```

```python
# Para exibir as linhas de índice 1 e 3

dados.loc[[1,3]]
```

```python
# Para filtrar dados

dados.loc[dados['purpose'] == 'radio/tv'] # Para filtrar as linhas em que os valores da coluna 'purpose' sejam iguais a 'radio/tv'
```

```python
# Filtrando dados utilizando outra condição

dados.loc[dados['credit_amount'] > 18000]
```

```python
# Podemos atribuir o resultado do filtro a uma variável. Dessa forma, criaremos outro dataframe

credito_2 = dados.loc[dados['credit_amount'] > 18000]
print(credito_2)
```

```python
# Definindo somente algumas colunas para um novo dataframe e utilizando um filtro

credito_3 = dados[['checking_status', 'duration']].loc[dados['credit_amount'] > 18000]
print(credito_3)
```

```python
# Séries = Única coluna
# Podem ser criadas a partir de listas, arrays do NumPy ou coluna de dataframe.

s1 = pd.Series([2, 5, 3, 34, 54, 23, 1, 16]) # Série a partir de uma lista
print(s1)
```

```python
# Série a partir de um array do NumPy

import numpy as np

array1 = np.array([2, 5, 3, 34, 54, 23, 1, 16])
s2 = pd.Series(array1)
print(s2)
```

```python
# Série a partir de um dataframe

s3 = dados['purpose']
print(s3)
print('-----')

# Dataframe similar a uma série

d4 = dados[['purpose']]

# Note a diferença

print(type(s3)) # Aqui temos uma série
print(type(d4)) # Aqui temos um dataframe
```

```python
# Para renomear colunas
# Utiliza-se a sintaxe de dicionários
# IMPORTANTE: essa alteração não é persisida

dados.rename(columns={'duration': 'duração', 'purpose': 'propósito'})
```

```python
# Observe que a alteração anterior não foi persistida

dados.head(1)
```

```python
# Para persistir a alteração

dados.rename(columns={'duration': 'duração', 'purpose': 'propósito'}, inplace = True)

# Por padrão, o parâmetro inplace = False
# O inplace = True garante que a alteração será persistida
```

```python
dados.head(1)
```

```python
# Para excluir uma coluna (e persistir a alteração)

dados.drop('checking_status', axis = 1, inplace = True)
print(dados)
```

```python
# Verificar dados nulos
# Essa alternativa não é funcial para grandes volumes de dados

dados.isnull()
```

```python
# Verificar dados nulos

dados.isnull().sum()
```

```python
# Retirar colunas com NaN

dados.dropna()
```

```python
# Preencher dados faltantes

dados['duração'].fillna(0, inplace = True)
```

```python
# iloc para exibir determinadas linhas e colunas

dados.iloc[0:3, 0:5]
```
